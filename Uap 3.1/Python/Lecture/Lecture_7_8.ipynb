Union

```python
set_a = {1, 2, 3}
set_b = {3, 4, 5}

# Using | operator
union_set = set_a | set_b

print(union_set)
```

```python
# Using union() method
set_a = {1, 2, 3}
set_b = {3, 4, 5}
union_set = set_b.union(set_a)
print(union_set)
```

 Intersection (& or .intersection())

```python
set_a = {1, 2, 3}
set_b = {3, 4, 5}

# Using & operator
intersection_set = set_a & set_b

# Using intersection() method
intersection_set = set_a.intersection(set_b)

print(intersection_set)
```

 Difference (- or .difference())

```python
set_a = {1, 2, 3}
set_b = {1, 2}

# Using - operator
difference_set = set_b - set_a

print(set_b - set_a == set_a - set_b)
print(difference_set)
```

```python
set_a = {1,2, 3}
set_b = {3}

# Using difference() method
difference_set = set_b.difference(set_a)
```

```python
print((set_a - set_b) == (set_b - set_a))
```

Symmetric Difference (^ or .symmetric_difference())

```python
set_a = {1, 2, 3}
set_b = {3, 4, 5}

# Using ^ operator
sym_diff_set = set_a ^ set_b

# Using symmetric_difference() method
#sym_diff_set = set_a.symmetric_difference(set_b)

print(sym_diff_set)
```

Subset and Superset

```python
a = {4}
b = {1, 2, 3, 4}

# Check if set_a is a subset of set_b
print(a <= b)

# Check if set_b is a superset of set_a
print(b >= a)  # Output: True
```

```python
def max_fun(list1):
  max_value = 0
  for i in list1:
    if i>max_value:
      max_value = i
  return max_value

list2 = [10,1,3,5,1,0]
print(max_fun(list2))
```

```python
def grade_count(marks):
  count = 0
  for i in marks:
    if i >= 80:
      count += 1
  return count

students_marks = [70,45,67,99,97,80]
print(grade_count(students_marks))
```

```python
def square_elements(nums):
    squared_numbers = []
    for num in nums:
        squared_numbers.append(num ** 2)
    return squared_numbers

my_numbers = [1, 2, 3, 4]
squared = square_elements(my_numbers)
print(squared)
```

```python
#pass multiple list in function
def merge_lists(list1, list2):
    return list1 + list2

list_a = [1, 2, 3]
list_b = [4, 5, 6]
combined = merge_lists(list_a, list_b)
print(combined)
```

```python
def concatenate_lists(*v):
  result = []
  for i in v:
    result.append(sum(i))

  return max(result)

list_1 = [80,50,69,21]
list_2 = [80,97.56,33]
list_3 = [80,59,60,21]
list_4 = [67,50,69,85]

summation = concatenate_lists(list_1, list_2, list_3,list_4)
print(summation)
```

```python
# Function that takes a tuple as an argument
def print_tuple(my_tuple):
    for item in my_tuple:
        print(item)

example_tuple = (1, 2, 3, 4, 5)

print_tuple(example_tuple)
```

```python
# Function that takes a dictionary as an argument

def print_dictionary(my_dict):
    for key, value in my_dict.items():
        print(key, ":", value)

example_dict = {"name": "Alice",
                "age": 30,
                "city": "New York"}

print_dictionary(example_dict)
```

```python

def calculate_marks(students):
    results = {}

    for s, m in students.items():
        print(s)
        print(m)
        total = sum(m)
        results[s] = total

    return results

students_marks = {
    'Alice': [85, 90, 78],
    'Bob': [70, 82, 88],
    'Charlie': [95, 85, 90],
}
marks_summary = calculate_marks(students_marks)
print(marks_summary)
```
