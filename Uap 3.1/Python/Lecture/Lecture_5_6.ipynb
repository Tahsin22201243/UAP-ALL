# **Tuple**


*   Tuples are **indexed** **ordered**, and **immutable**.
*   Tuple can have duplicates.



```python
students = ('david', 'john', 'mary')
print(students)
print(type(students))
```

```python
print(students[0])
```

```python
for student in students:
  print(student)
```

```python
students[1] = 'Jemmy'
```

```python
students.append('Jemmy')
```

```python
students.remove('john')
```

```python
updated_student = students +('jemmy',) #mandatory to use the comma after one item
```

```python
print(updated_student)
```

```python
print(students)
```

```python
my_tuple = (1, 2, 3)
print(my_tuple)
a, *b= my_tuple
print(a)
print(b)
```

```python
a, *b = my_tuple
print(a)
print(b)
```

## **Add, change value, remove, and insert in a tuple using list conversion.**

```python
fruits = ('Apple', 'Banana', 'Orange', 'Grapes')

fruit_list = list(fruits)

fruit_list.append('Mango')

fruit_list.insert(2, 'Strawberry')

my_fruits = tuple(fruit_list)

print(fruits)
print(my_fruits)
```

```python
fruits = ('Apple', 'Banana', 'Orange', 'Grapes')

fruit_list = list(fruits)

fruit_list.remove('Orange')

fruits = tuple(fruit_list)

print(fruits)
```

```python
#take input in a tuple using for loop

n = int(input("Enter the number of elements you want in the tuple: "))

input_list = []

for i in range(n):
    element = input("Enter element: ")
    input_list.append(element)

input_tuple = tuple(input_list)

print("The tuple is:", input_tuple)
```

# **Dictionary**


```python
students = {'22101011': 'Urmi',
            '22101012': 'Hasib',
            '22101013': 'Mamun'}
print(students)
print(type(students))
```

```python
print(students['22101011'])
```

```python
students['22101011'] = 'Akash'
```

```python
print(students['Akash'])
```

```python
print(students)
```

```python
students.get('22101011')
```

```python
print(students.keys())
print(students.values())
print(students.items())
```

```python
students['22101014'] = 'Nabila'
print(students)
```

```python
students.update({'22101011': 'Peter'})
students['22101020'] = 'Sajib'
print(students)
```

```python
print(students.pop('22101011'))
print(students)
```

```python
item = students.popitem()
print(item)
print(students)
```

```python
students.clear()
print(students)
```

```python
del students
print(students)
```

```python
students = {'22101011': 'Fahim',
            '22101012': 'Borhan',
            '22101013': 'Mostak'}


students_copy = students
print(students_copy)
```

```python
students_copy['22101011'] = 'Peter'
print(students_copy)
print(students)
```

```python
# prompt: COPY OF dictionary

students = {'22101011': 'David',
            '22101012': 'John',
            '22101013': 'Mary'}

# Method 1: Using the copy() method
students2 = students.copy()
print(students2)

# Method 2: Using dict() constructor
students3 = dict(students)
print(students3)
```

```python
students2['22101011'] = 'Jemmy'
print(students2)
print(students)
```
