## **Class variable**

```python
class Player:
  total_run = 0
  def __init__(self, name, run):
    self.name = name
    self.run = run


  def hit_four(self):
    self.run += 4
    Player.total_run += 4

  def hit_six(self):
    self.run += 6
    Player.total_run += 6

sadman = Player("Sadman", 0)
print(sadman.run)
sadman.hit_four()
sadman.hit_six()
print(sadman.run)
borhan = Player("Borhan", 0)
print(borhan.run)
borhan.hit_four()
print(borhan.run)
print(borhan.total_run)
print(Player.total_run)

```

The University class keeps track of the total number of students across all departments. Initially, four departments are created with the following student counts: 120 in CSE, 80 in BBA, 50 in English, and 70 in Civil. However, no new students have been admitted yet.

```python
class University:
    total_students = 0 #class variable

    def __init__(self, department, student_count):
        self.department = department
        self.student_count = student_count
        University.total_students += student_count


    def admit_new_students(self,no_of_newstudent):
        self.student_count += no_of_newstudent
        University.total_students += no_of_newstudent

cse_dept = University("CSE", 120)
bba_dept = University("BBA", 80)
english_dept = University("English", 50)
civil_dept = University("Civil", 50)

print(cse_dept.student_count)

print("Total number of students in the university:", University.total_students)
```

```python
class LibraryBook:
    total_books_borrowed = 0

    def __init__(self, title, author):
        self.title = title
        self.author = author
        self.is_borrowed = False

    def borrow_book(self):
        if not self.is_borrowed:
            self.is_borrowed = True
            LibraryBook.total_books_borrowed += 1
            print(self.title, 'has been borrowed.')
        else:
            print(self.title, 'is already borrowed.')

    def return_book(self):
        if self.is_borrowed:
            self.is_borrowed = False
            LibraryBook.total_books_borrowed -= 1
            print(self.title, 'has been returned.')
        else:
            print(self.title, 'was not borrowed.')


book1 = LibraryBook("A", "x")
book2 = LibraryBook("B", "y")
book3 = LibraryBook("C", "z")
book1.borrow_book()
book2.borrow_book()
book2.return_book()
print("Total books borrowed from the library:", LibraryBook.total_books_borrowed)
```

```python
class Employee:
    total_attendance = 0

    def __init__(self, name):
        self.name = name
        self.attendance_marked = False

    def mark_attendance(self):
        if not self.attendance_marked:
            self.attendance_marked = True
            Employee.total_attendance += 1
            print(f"Attendance marked for {self.name}.")
        else:
            print(f"{self.name} has already marked attendance.")



emp1 = Employee("Alice")
emp2 = Employee("Bob")
emp3 = Employee("Charlie")
emp1.mark_attendance()
emp2.mark_attendance()

print("Total attendance for the day:", Employee.total_attendance)
```

```python
class Product:

    total_quantity = 0

    def __init__(self, name, quantity):
        self.name = name
        self.quantity = quantity
        Product.total_quantity += quantity

    def add_stock(self, quantity):
        self.quantity += quantity
        Product.total_quantity += quantity

    def sell_product(self, quantity):
        if quantity > self.quantity:
            print(f"Not enough stock of {self.name} to sell.")
        else:
            self.quantity -= quantity
            Product.total_quantity -= quantity
            print(f"{quantity} units of {self.name} sold.")

    @classmethod
    def get_total_quantity(cls):
        return cls.total_quantity

product1 = Product("Laptop", 50)
product2 = Product("Smartphone", 100)
product3 = Product("Headphones", 75)

product1.sell_product(10)
product2.sell_product(20)

print("Total quantity of all products in inventory:", Product.get_total_quantity())
```

## **Class method**

```python
class University:
    university_name = "UAP"

    def __init__(self, department, student_count):
        self.department = department
        self.student_count = student_count


    @classmethod
    def update_university_name(cls, new_name):
        cls.university_name = new_name

    def display_info(self):
        print(f"University: {University.university_name}")
        print(f"Department: {self.department}")
        print(f"Student Count: {self.student_count}")



cse_dept = University("CSE", 120)
bba_dept = University("BBA", 80)
cse_dept.display_info()
bba_dept.display_info()
University.update_university_name("ABC")
cse_dept.display_info()
bba_dept.display_info()


```
