# why we use self in constructor

The code you provided demonstrates the use of 'self' in a Python class.
'self' is a convention used in Python to refer to the instance of the class itself.
It is a way of accessing the attributes and methods of the class within the class definition.
In the constructor (`__init__` method), we use 'self' to bind attributes like name and breed to the specific instance of the class being created.

Example
When we create an instance like:
my_dog = Dog("Buddy", "Golden Retriever")
The constructor Dog.__init__(my_dog, "Buddy", "Golden Retriever") is called.
Inside the constructor:
self.name = name assigns the value "Buddy" to the name attribute of the 'my_dog' instance.
self.breed = breed assigns the value "Golden Retriever" to the breed attribute of the 'my_dog' instance.

Without self, the assignment would not be bound to the specific instance being created.

 In essence, 'self' is used to ensure that we are working with the correct instance of the class when we access and modify attributes and call methods.

```python
# prompt: example of class and object

class Car:
  def __init__(self, make, model, year):
    self.make = make
    self.model = model
    self.year = year

  def start_engine(self):
    print(f"The {self.year} {self.make} {self.model}'s engine is starting.")


my_car = Car("Toyota", "Camry", 2023)
your_car = Car("Honda", "Civic", 2022)

print(f"My car is a {my_car.make} {my_car.model}.")
print(f"Your car is a {your_car.make} {your_car.model}.")

my_car.start_engine()
your_car.start_engine()
```

Scenario 1: Book Class
Create a class named Book that represents a book. The class should have:

An __init__ method that takes in the book's title and author.
A method called get_info that prints the title and author of the book.
Create an object of this class and use it to print the information of a book titled "The Great Gatsby" by F. Scott Fitzgerald.

```python
class Book:
  def __init__(self, title, author):
    self.title = title
    self.author = author

  def get_info(self):
    print(f"{self.title} by {self.author}")
    print(self.title, ' by', self.author)

book_1 = Book("The Great Gatsby", "F. Scott Fitzgerald")
book_1.get_info()
```

```python

class Animal:
    def __init__(self, name, species):
        self.name = name
        self.species = species

    def make_sound(self):
        print("Generic animal sound")


class Dog(Animal):
    def __init__(self, name, breed):
        super().__init__(name, species="Dog")
        self.breed = breed



my_dog = Dog("Buddy", "Golden Retriever")
print(my_dog.name)
print(my_dog.species)
print(my_dog.breed)
my_dog.make_sound()
```

```python

class Animal:
    def __init__(self, name):
        self.name = name

    def eat(self):
        print(f"{self.name} is eating.")


class Mammal(Animal):
    def __init__(self, name, fur_color):
        super().__init__(name)
        self.fur_color = fur_color

    def give_birth(self):
        print(f"{self.name} is giving birth.")


class Dog(Mammal):
    def __init__(self, name, fur_color, breed):
        super().__init__(name, fur_color)
        self.breed = breed

    def bark(self):
        print(f"{self.name} is barking.")


my_dog = Dog("Buddy", "Golden", "Golden Retriever")
print(my_dog.name)  # Output: Buddy
print(my_dog.fur_color)  # Output: Golden
print(my_dog.breed)  # Output: Golden Retriever

my_dog.eat()  # Output: Buddy is eating.
my_dog.give_birth()  # Output: Buddy is giving birth.
my_dog.bark()  # Output: Buddy is barking.
```
