# **Method overloading**

```python
class Calculator:
  def add(self, num1,num2):
    print(num1+num2)
    print("from method 1")

  def add(self, num1,num2,num3):
    print(num1+num2+num3)
    print("from method 2")

c1 = Calculator()
c1.add(1,2)
c1.add(1,2,3)
```

```python
class Calculator:#Using Default Arguments
  def add(self, num1=0,num2=0,num3=0):
    print(num1+num2)


c = Calculator()
c.add(2,2)
```

```python
class Calculator:#Using Variable-Length Arguments
  def add(self, *v):
    return sum(v)

c = Calculator()
c.add(1,2,3,6,7)
```

```python
#topic: method overloading
class Calculator:
  def add(self, num1=0,num2=0,num3=0):
    print(num1+num2+num3)
    print('this is from first method')

  def add(self, *v):
    print('this is from second method')
    return sum(v)

c1 = Calculator()
c1.add(1)
c1.add(1,2)
c1.add(1,2,3)
```

```python
from multipledispatch import dispatch


class Calculator:

  @dispatch(int,int)
  def add(self, num1,num2):
    print(num1+num2)
    print("from method 1")

  @dispatch(int,int,int)
  def add(self, num1,num2,num3):
    print(num1+num2+num3)
    print("from method 2")

  @dispatch(float,float)
  def add(self, num1,num2):
    print(num1+num2)
    print("from method 3")

  @dispatch(int,float)
  def add(self, num1,num2):
    print(num1+num2)
    print("from method 4")

c1 = Calculator()
c1.add(1,2)
c1.add(1,2,3)
c1.add(1.2,2.4)
c1.add(2, 3.5)
```

# **Inheritance**

```python
class Parent:
    def parent_method(self):
        print("This is the parent method.")

class Child(Parent):
    def child_method(self):
        print("This is the child method.")

c = Child()
c.parent_method()
c.child_method()
p = Parent()
p.parent_method()
p.child_method()
```

```python
class Animal:
    def __init__(self, name,a):
        self.name = name
        self.a = a
        print('from parent constructor')

    def speak(self):
        print(self.name, 'makes a sound.')


class Dog(Animal):
    def __init__(self, name,a, breed):
        super().__init__(name,a)
        self.breed = breed
        print("In child constructor")


    def bark(self):
        ##print(self.name, "the",self.breed, "barks!")
        print('anything')

    def update_breed(self, new_breed):
        self.breed = new_breed
        print(self.breed)



d = Dog('d1', 12, 'Labrador')
d.speak()
d.bark()
d.update_breed('b2')

```

A Parent class Vehicle has a method start() that prints "Vehicle started." The Child class Car inherits from Vehicle and can call start() to print "Vehicle started" while having its own method drive() to print "Car is driving.

```python
class Vehicle:
    def start(self):
        print("Vehicle started.")

class Car(Vehicle):
    def drive(self):
        print("Car is driving.")

car = Car()
car.start()
car.drive()
```

```python

class Father:
    def __init__(self, father_name):
        self.father_name = father_name

    def details(self):
        print("From Father class")

class Mother:
    def __init__(self, mother_name):
        self.mother_name = mother_name

    def details(self):
        print("From mother class")

class Child(Mother, Father):
    def __init__(self, father_name, mother_name, child_name):
        Father.__init__(self, father_name)
        Mother.__init__(self, mother_name)
        self.child_name = child_name

    def show_child(self):
        print(f"Child: {self.child_name}")


child = Child("John", "Jane", "Emily")
child.details()
```

```python
class Father:
    def __init__(self, father_name):
        self.father_name = father_name

    def show_father(self):
        print(f"Father: {self.father_name}")

class Mother:
    def __init__(self, mother_name):
        self.mother_name = mother_name

    def show_mother(self):
        print(f"Mother: {self.mother_name}")

class Child(Father, Mother):
    def __init__(self, father_name, mother_name, child_name):
        Father.__init__(self, father_name)
        Mother.__init__(self, mother_name)
        self.child_name = child_name

    def show_child(self):
        print(f"Child: {self.child_name}")


child = Child("John", "Jane", "Emily")
child.show_father()
child.show_mother()
child.show_child()
```
