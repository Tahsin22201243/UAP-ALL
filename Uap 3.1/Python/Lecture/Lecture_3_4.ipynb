1. Write a Python program that takes an integer input from the user and checks whether the number is odd or even.

2.Write a program that takes an integer input and prints whether the number is positive, negative, or zero.

3.Write a Python program that takes the lengths of three sides of a triangle as input and determines whether it is an equilateral, isosceles, or scalene triangle.

  

```python
a = int(input("Enter a number: "))

if(a%2 == 0):
  print("Even")
else:
  print("Odd")
```

```python
a = int(input("Enter a number: "))
if(a>0):
  print("Positive")
elif(a<0):
  print("Negative")
else:
  print("Zero")
```

```python
first_side = int(input("Enter first side: "))
second_side = int(input("Enter second side: "))
third_side = int(input("Enter third side:"))

if(first_side == second_side == third_side):
  print("Equilateral")
elif(first_side != second_side!= third_side):
  print("Scalene")
else:
  print("Isosceles")
```

```python
#take n number of input in a list
n = int(input("Enter the number of elements: "))  # Number of elements in the list
my_list = []
sum = 0
for i in range(n):
    element = int(input("Enter element : "))
    my_list.append(element)
    sum += element

print("The list is:", my_list)
print(f"The sum is {sum}")
```

```python
numbers = [1, 2, 3, 4, 5]
squares = []
for number in numbers:
    squares.append(number ** 2)

# List comprehension version
squares = [number ** 2 for number in numbers]
print(squares)
```
